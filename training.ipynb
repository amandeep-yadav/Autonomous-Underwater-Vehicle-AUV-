```python
import shutil
import zipfile
# Path to the folder you want to zip
folder_to_zip = '/kaggle/working/yolov5/runs/detect/exp'

# Name of the zip file
zip_filename = '/kaggle/working/my_folder1.zip'

# Create a zip file
with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
    # Walk through the folder and add each file to the zip file
    for foldername, subfolders, filenames in os.walk(folder_to_zip):
        for filename in filenames:
            # Get the full path of the file
            file_path = os.path.join(foldername, filename)
            # Add the file to the zip file
            zipf.write(file_path, os.path.relpath(file_path, folder_to_zip))

print(f"Folder '{folder_to_zip}' zipped successfully to '{zip_filename}'.")

# Name of the zip file
zip_filename = '/kaggle/working/my_folder.zip'

# Create a zip file
with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
    # Walk through the folder and add each file to the zip file
    for foldername, subfolders, filenames in os.walk(folder_to_zip):
        for filename in filenames:
            # Get the full path of the file
            file_path = os.path.join(foldername, filename)
            # Add the file to the zip file
            zipf.write(file_path, os.path.relpath(file_path, folder_to_zip))

print(f"Folder '{folder_to_zip}' zipped successfully to '{zip_filename}'.")
```
